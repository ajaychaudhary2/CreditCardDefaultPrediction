```python
import pandas as pd
```

```python
data = pd.read_csv(r"E:\Data_Science _master\ML\prac_project\Credit Card Default Prediction\notebooks\data\UCI_Credit_Card.csv")
```

```python
columns_to_drop = ["ID", "BILL_AMT4", "BILL_AMT5", "BILL_AMT6"]
data = data.drop(columns=columns_to_drop)
```

```python
duplicates_count = data.duplicated().sum()
print(f"Number of duplicates: {duplicates_count}")
```

```python
# Drop duplicates
data = data.drop_duplicates()

# Verify removal
print(f"Data shape after removing duplicates: {data.shape}")
```

```python
 from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.25 , random_state=42)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Define a pipeline for numerical data
num_pipeline = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="mean")),  # Impute missing values with the mean
        ("scaler", StandardScaler()),                # Standardize numerical features
    ]
)

# Fit and transform your numerical data
X_scaled = num_pipeline.fit_transform(X)  # X is your input numerical DataFrame
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt', 'log2'],
    'bootstrap': [True, False]
}
```

```python
random_search = RandomizedSearchCV(estimator=RandomForestClassifier(random_state=42),
                                   param_distributions=param_grid,
                                   n_iter=20,  # Number of random combinations to sample
                                   cv=5,       # Cross-validation folds
                                   n_jobs=-1,   # Use all cores
                                   verbose=2,   # Verbosity for progress
                                   random_state=42, 
                                   scoring='accuracy')
```

```python
random_search.fit(X_train, y_train)
```

```python
print(f"Best Hyperparameters: {random_search.best_params_}")
best_model = random_search.best_estimator_
```

```python

# Make predictions with the best model
y_pred = best_model.predict(X_test)
```

```python
# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Classification Report
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Confusion Matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))   
```
